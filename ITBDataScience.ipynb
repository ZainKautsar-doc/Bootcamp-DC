## Corelation Variable

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(),annot=True, cmap='coolwarm')
plt.title('Corelation Variable')
plt.savefig('Corelation Variable.png')
```

## Kesimpulan data


*   Age -> HeartDisease : (0,28), semakin tinggi atau semakin tua umur seorang, kemungkinan besar terkena penyakit jantung
*   RestingBP -> HeartDisease : (0.11), semakin tinggi tekanan darah saat istirahat,  kemungkinan besar akan semakin tinggi terkena penyakit jantung
*   Cholestrol ->  HeartDisease (-0.23), semakin tinnggi kadar cholestrol, maka kemungkinan terkena penyakit jantung
*   FastingBS -> HeartDisease (0.27), jika semakin tinggi gula darah, maka berpotensi mendapat penyakit jantung karena organ yang gampang terkena penyakit jantung
*   MaxHR -> HeartDisease (-0.4), jika semakin tinggi detak jantung, maka jantung aman atau normal, jika detak jantung lemah atau rendah menandakan jantung yg lemah
*   OldPeak -> HeartDisease (0.4), jika semakin tinggi seseorang depresi makan akan mengganggu kerja jantung dan akan semakin rentan untuk terkena serangan jantung
